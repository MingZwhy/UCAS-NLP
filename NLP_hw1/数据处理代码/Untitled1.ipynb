```python
import math

def handle_files(path_list,max_num):
    # 统计每个汉字出现的次数
    count = 0
    word_dict = {}
    for path in path_list:
        files = glob.glob(path + '\\*.txt')
        for file in tqdm(files):
            with open(file, 'r', encoding='utf-8') as f:
                text = f.read()
            for char in text:
                if '\u4e00' <= char <= '\u9fff':
                    if char in word_dict:
                        word_dict[char] += 1
                    else:
                        word_dict[char] = 1
                    count = count + 1
                if count >= max_num:
                    max_20word(word_dict, count)
                    return cal_entropy(word_dict, count)
    assert(False)
```

```python
def cal_entropy(word_dict, count):
    H = 0
    for word_value in word_dict.values():
        p = word_value / count
        H -= p * math.log2(p)
    return H
```

```python
def max_20word(word_dict,count):
    sort_dict = sorted(word_dict.items(), key=lambda x: x[1], reverse=True)
    for key, value in sort_dict[:20]:
        print(key)
        print(value)
```
