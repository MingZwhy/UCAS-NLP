```python
import string
import re
#消除标点符号（包括中英文）
def filter_punctuation(words): 
    new_words = []; 
    illegal_char = string.punctuation + '【·！…（）—：“”？《》、；】' 		
    pattern=re.compile('[%s]' % re.escape(illegal_char)) 
    for word in words: 
        new_word = pattern.sub(u'', word) 
        if not new_word == u'': 
            new_words.append(new_word) 
    return new_words

#去停用词
def handleStopWords(words):
    stops = set(nltk.corpus.stopwords.words('english'))
    words = [word for word in words if word.lower() not in stops]
    return words
```

```python
vector_len = 5

dataset = []

for (i,word) in enumerate(num_text):
    if (word != 0):
        if (i > vector_len):
            seq = num_text[i-vector_len : i+1]
            dataset.append(seq)
        elif (i < vector_len):
            seq = []
            left = vector_len - i
            while(left > 0):
                seq.append(0)
                left = left - 1
                
            seq = seq + num_text[:i+1]
            dataset.append(seq)
        else:
            seq = num_text[:i+1]
            dataset.append(seq)
```

```python
lstm_embedding_result = {}

for word in word_index.keys():
    index = word_index[word]
    array = lstm_weights[0][index]
    np_array = np.array(array)
    lstm_embedding_result[word] = np_array
```

```python
lstm_distance = np.zeros((num_keys + 1 , num_keys + 1))
lstm_distance.shape
```

```python
for i in range(num_keys + 1):
    word1 = list(word_index.keys())[i]
    array1 = lstm_embedding_result[word1]
    for j in range(num_keys + 1):
        word2 = list(word_index.keys())[j]
        array2 = lstm_embedding_result[word2]
        #计算欧式距离
        distance = np.linalg.norm(array1 - array2)
        lstm_distance[i][j] = distance
```

```python
#计算与某个词最近的10个词
likely_words_num = []
likely_words_word = {}
for i in range(num_keys + 1):
    array = np.array(lstm_distance[i])
    #sorted_list = np.argsort(array)
    sorted_list = list(np.argsort(array))
    sorted_list_10 = sorted_list[1:11]
    likely_words_num.append(sorted_list_10)
    
    words = []
    for j in range(len(sorted_list_10)):
        word = index_word[sorted_list_10[j]]
        words.append(word)
    likely_words_word[(list(word_index.keys()))[i]] = words
    
```
